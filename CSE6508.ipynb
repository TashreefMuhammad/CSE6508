```python
full_data['ma5'] = talib.SMA(full_data['Close'].values, timeperiod=5)
full_data['ma15'] = talib.SMA(full_data['Close'].values, timeperiod=15)
full_data['ma25'] = talib.SMA(full_data['Close'].values, timeperiod=25)
full_data['ma150'] = talib.SMA(full_data['Close'].values, timeperiod=150)
full_data['ma200'] = talib.SMA(full_data['Close'].values, timeperiod=200)
full_data['ma300'] = talib.SMA(full_data['Close'].values, timeperiod=300)
full_data['ma10'] = talib.SMA(full_data['Close'].values, timeperiod=10)
full_data['ma20'] = talib.SMA(full_data['Close'].values, timeperiod=20)
full_data['ma100'] = talib.SMA(full_data['Close'].values, timeperiod=100)
full_data['ma50'] = talib.SMA(full_data['Close'].values, timeperiod=50)
full_data['rsi14'] = talib.RSI(full_data['Close'].values, timeperiod=14)
full_data['ema20'] = talib.EMA(full_data['Close'].values, timeperiod=20)
full_data['macd(12,26)'], full_data['macdSl'], full_data['macdHist'] = talib.MACD(full_data['Close'].values, fastperiod=12, slowperiod=26, signalperiod=9)
full_data['%k'], full_data['%d'] = talib.STOCH(full_data['High'], full_data['Low'], full_data['Close'], fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
full_data['%R'] = talib.WILLR(full_data['High'], full_data['Low'], full_data['Close'], timeperiod=14)
full_data['upperband'], full_data['middleband'], full_data['lowerband'] = talib.BBANDS(full_data['Close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
full_data['ROC'] = talib.ROC(full_data['Close'], timeperiod = 10)
full_data['ADX'] = talib.ADX(full_data['High'], full_data['Low'], full_data['Close'], timeperiod=14)
full_data['ADXR'] = talib.ADXR(full_data['High'], full_data['Low'], full_data['Close'], timeperiod=14)
full_data['CCI'] = talib.CCI(full_data['High'], full_data['Low'], full_data['Close'], timeperiod=14)

indicators = list(full_data.columns)

full_data
```

```python
full_data.dropna(how='any', axis=0, inplace=True)

full_data
```

```python
full_data = full_data.tail(250)

full_data
```

```python
def simulateSolution(solution):

  reservedCapital = 10000
  reservedStock = 0
  current = reservedCapital

  # print(solution)

  for index, row in full_data.iterrows():
    buy = 1
    sell = 1

    for j in range(1, 10, 2):
      if row[indicators[solution[j - 1] + 6]] <= row[indicators[solution[j] + 6]]:
        buy = -1
        break
    for j in range(11, 20, 2):
      if row[indicators[solution[j - 1] + 6]] <= row[indicators[solution[j] + 6]]:
        sell = -1
        break
    if buy == 1:
      # buy condition has meet
      toBuyAmount = int(0.5 * reservedCapital)
      toBuyStock = int(toBuyAmount / row['Close'])
      toBuyAmount = int(toBuyStock * row['Close'])
      reservedCapital -= toBuyAmount
      reservedStock += toBuyStock
    elif sell == 1:
      # sell condition has meet
      toSellStock = int(0.8 * reservedStock)
      toSellAmount = toSellStock * row['Close']
      reservedCapital += toSellAmount
      reservedStock -= toSellStock
    current = reservedCapital + reservedStock * row['Close']
  return current

def fitness_func(ga_instance, solution, solution_idx):
  fitness = simulateSolution(solution)

  return fitness
```

```python
fitness_function = fitness_func

num_generations = 50
num_parents_mating = 10

sol_per_pop = 1000
num_genes = 20

init_range_low = 0
init_range_high = len(indicators) - 1 - 6

parent_selection_type = "sss"
keep_parents = 1

crossover_type = "single_point"

mutation_type = "random"
mutation_percent_genes = 10
```

```python
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       gene_type = int,
                       gene_space = range(0, len(indicators) - 1 - 6),
                       init_range_low=init_range_low,
                       init_range_high=init_range_high,
                       parent_selection_type=parent_selection_type,
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes
                       )

ga_instance.run()

ga_instance.plot_fitness()
```

```python
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))

print('Buy Condition')
for i in range(1, 11, 2):
  print(indicators[solution[i - 1] + 6], '>', indicators[solution[i] + 6])

print('Sell Condition')
for i in range(11, 20, 2):
  print(indicators[solution[i - 1] + 6], '>', indicators[solution[i] + 6])
```
